```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

$$ y = mx + b $$


The least squares method in linear regression minimizes the sum of the squared differences between the observed $$y_i $$ and predicted $$mx_i + b$$ values. Mathematically, it involves minimizing the expression

$$ \sum_{i=1}^{n} (y_i - (mx_i + b))^2 $$

where `n` is the number of data points,  $$ x_i, y_i $$ are the coordinates of each data point, `m` is the slope of the line, and `b` is the y-intercept. The method seeks to find the values of `m` and `b` that best fit the given data by minimizing the sum of the squared residuals, representing the vertical deviations between the observed and predicted values along the regression line.


```python
# Training the Simple Linear Regression model on the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Assuming y_train, y_test, and regressor are available
y_pred_train = regressor.predict(X_train)
y_pred_test = regressor.predict(X_test)

# Calculate MSE and RMSE for the training set
mse_train = np.mean((y_train - y_pred_train)**2)
rmse_train = np.sqrt(mse_train)

# Calculate MSE and RMSE for the test set
mse_test = np.mean((y_test - y_pred_test)**2)
rmse_test = np.sqrt(mse_test)

print(f"MSE on Training Set: {mse_train}")
print(f"RMSE on Training Set: {rmse_train}")
print(f"MSE on Test Set: {mse_test}")
print(f"RMSE on Test Set: {rmse_test}")
```

Mean Squared Error (MSE) is a measure of the average squared difference between the actual $$y_i\ $$ and predicted $$ \hat{y}_i\ $$values:

$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 $$

Root Mean Squared Error (RMSE) is the square root of the MSE, providing a more interpretable measure of the average magnitude of prediction errors:

$$ \text{RMSE} = \sqrt{\text{MSE}} $$

In these equations, `n` represents the number of data points. Lower MSE and RMSE values indicate better model performance.


a good value for Mean Squared Error `MSE` and Root Mean Squared Error `RMSE` is one that is low, indicating minimal discrepancy between predicted and actual values. MSE represents the average squared differences, and RMSE, being its square root, provides a more interpretable metric in the original units of the dependent variable. Lower values of MSE and RMSE suggest better model accuracy, with an ideal scenario being close to zero, signifying precise predictions.

```python
import matplotlib.pyplot as plt

# Scatter plot for training set
plt.scatter(y_train, y_pred_train, color='blue')
plt.title('Actual vs Predicted (Training Set)')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()

# Scatter plot for test set
plt.scatter(y_test, y_pred_test, color='green')
plt.title('Actual vs Predicted (Test Set)')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.show()
```

R-squared `coefficient of determination` is a metric that quantifies the proportion of the variance in the dependent variable that is explained by the independent variables in the model. It is a scale-independent measure that ranges from 0 to 1, with higher values indicating a better fit of the model to the data.

```python
from sklearn.metrics import r2_score

r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)

print(f"R-squared on Training Set: {r2_train}")
print(f"R-squared on Test Set: {r2_test}")
```

The R-squared (coefficient of determination) is calculated using the following formula:


**$$ R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2} $$**

where $$y_i$$ represents the actual values, $$ \hat{y}_i\ $$ represents the predicted values from the regression model, $$ \bar{y}\ $$ is the mean of the actual values, and `n` is the number of data points. This formula assesses the proportion of the total variance in the dependent variable explained by the regression model. An R-squared value closer to 1 implies a better fit of the model to the data.
